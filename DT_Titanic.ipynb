```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
# Load the Titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
# Drop unnecessary columns and handle missing values
titanic = titanic.drop(['deck', 'embark_town', 'alive', 'who', 'alone', 'adult_male'], axis=1)
titanic = titanic.dropna()
```

```python
# Convert categorical variables into numeric using one-hot encoding
titanic = pd.get_dummies(titanic, columns=['sex', 'embarked', 'class'], drop_first=True)
```

```python
# Define the features (X) and the target (y)
X = titanic.drop('survived', axis=1)
y = titanic['survived']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train the Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
# Evaluate the model on the test set
accuracy = clf.score(X_test, y_test)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Visualize the Decision Tree using plot_tree
plt.figure(figsize=(100, 70))
plot_tree(clf, feature_names=X.columns, class_names=['Not Survived', 'Survived'], filled=True, rounded=True)
plt.show()
```
