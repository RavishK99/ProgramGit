```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
# Load the Heart Disease dataset
data = pd.read_csv('heart.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# If needed, convert categorical columns to numerical values (no categorical variables in this dataset)
# You can encode categorical values here if necessary.

# Define the features (X) and the target (y)
X = data.drop('censors', axis=1)
y = data['censors']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train the Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
# Predict on test data
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Evaluate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Visualize the Decision Tree
plt.figure(figsize=(20, 10))
plot_tree(clf, feature_names=X.columns, class_names=['No Disease', 'Disease'], filled=True, rounded=True)
plt.show()
```
